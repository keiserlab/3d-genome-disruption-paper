## Deletions

### 5kb Deletions

Tiled 5kb deletions across the genome.

```python
runF = pd.read_csv('data/5kb_deletions.csv')
runF.head()
```

```python
plt.figure(figsize=(6,1), dpi=600)
ax = sns.histplot(runF['disruption'], color='white', linewidth=.2, bins=100)
plt.xlabel('Disruption [log(MSE)]')
plt.ylabel('')
ax.set_yticks([])
plt.yticks(color='w')
```
